```python
# print(arr.reshape(256,8).shape)
r=10
files_test=files

for file_ in files_test:
    pid = file_.parent.stem          # person id
    sid = file_.stem.split("_")[0]   # sample id
    # if (pid,sid) in sd:
        # continue
    out_ = base_ / pid
    out_.mkdir(exist_ok=True)
    data=file_.read_text().strip()
    try:
        arr = np.fromiter((int(c) for c in data), dtype=np.uint8)
    except ValueError as e:
        print(f'Error: {file_}')
        continue

    ic = arr.reshape(8,256)
    for shift in range(-r, r+1):
        rid=f"{shift:+}" if shift else '0'        # rot id
        out_file = f'{pid}_{sid}_{rid}.txt'
        # if shift==0:
            # p_ic(ic,)
        out_file_ = out_ / out_file
        if not out_file_.exists():
            pic = np.roll(ic,shift=shift) if shift else ic
            p_ic(pic,out_file_)
```
