```python
# combine_plot_results.py
import numpy as np
import plotly.graph_objects as go

# Example load - change to your actual filenames/formats
# Each file contains two arrays of same length: penetration_rates and hit_rates
hnsw = np.load('CASIAv1_hnsw.npz')   # keys: 'x','y'
khnsw = np.load('CASIAv1_khnsw.npz')
lsh = np.load('CASIAv1_LSH.npz')
ivf = np.load('CASIAv1_IVF.npz')

datasets = {
    'HNSW' : hnsw,
    'KHNSW': khnsw,
    'LSH'  : lsh,
    'IVF'  : ivf
}

# choose markers and line styles for clarity
style = {
    'HNSW': {'marker':'circle',  'dash':'solid'},
    'KHNSW':{'marker':'square',  'dash':'dash'},
    'LSH':  {'marker':'diamond', 'dash':'dot'},
    'IVF':  {'marker':'cross',   'dash':'longdash'}
}

fig = go.Figure()
for name, data in datasets.items():
    x = data['x']  # Top K Retrieved ( Top_k/total Indexed )s
    y = data['y']  # hit-rates
    fig.add_trace(go.Scatter(
        x = x,
        y = y,
        mode = 'lines+markers',
        name = name,
        marker = dict(symbol=style[name]['marker'], size=8),
        line = dict(dash=style[name]['dash'], width=2),
        hovertemplate = f'{name}<br>Penetration: %{{x}}<br>Hit-rate: %{{y:.3f}}<extra></extra>'
    ))

fig.update_layout(
    title = 'CASIAv1',
    xaxis_title = 'Top K Retrieved ( Top_k/total Indexed )',
    yaxis_title = 'Hit-rate',
    legend = dict(title='Method', orientation='h', yanchor='bottom', y=1.02, xanchor='left', x=0),
    template = 'plotly_white',
    width = 900, height = 520
)

fig.write_html('combined_hitrate.html', include_plotlyjs='cdn')  # save interactive file
fig.show()
```

```python
# combine_plot_results.py
import numpy as np
import plotly.graph_objects as go

# Load only HNSW and KHNSW result files
hnsw = np.load('CASIAv1_hnsw.npz')   # keys: 'x','y'
khnsw = np.load('CASIAv1_khnsw.npz')

# Keep explicit order so KHNSW is added last (drawn on top)
ordered = [
    ('HNSW', hnsw),
    ('KHNSW', khnsw)
]

# Styles (opacity moved to trace-level usage)
style = {
    'HNSW':  {'marker':'circle',  'dash':'solid', 'color':'blue', 'line_width':2.5, 'opacity':0.6},
    'KHNSW': {'marker':'square',  'dash':'dash',  'color':'red',  'line_width':4.0, 'opacity':0.9}
}

fig = go.Figure()

for name, data in ordered:
    x = data['x']
    y = data['y']
    fig.add_trace(go.Scatter(
        x = x,
        y = y,
        mode = 'lines+markers',
        name = name,
        opacity = style[name]['opacity'],   # <-- trace-level opacity
        marker = dict(
            symbol = style[name]['marker'],
            size = 9,
            color = style[name]['color'],
            line = dict(width=1.5, color='black')  # outline
        ),
        line = dict(
            dash = style[name]['dash'],
            width = style[name]['line_width'],
            color = style[name]['color']
        ),
        hovertemplate = f'{name}<br>Penetration: %{{x}}<br>Hit-rate: %{{y:.3f}}<extra></extra>'
    ))

fig.update_layout(
    title = 'CASIAv1: HNSW vs KHNSW',
    xaxis_title = 'Top K Retrieved ( Top_k/total Indexed )',
    yaxis_title = 'Hit-rate',
    legend = dict(title='Method', orientation='h', yanchor='bottom', y=1.02, xanchor='left', x=0),
    template = 'plotly_white',
    width = 900, height = 520
)

fig.write_html('combined_hitrate_hnsw_khnsw.html', include_plotlyjs='cdn')
fig.show()
```
